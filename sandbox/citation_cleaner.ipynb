## Cleanup citation files
Trim the various bib files.

```python
bibfilepath = "../ms_files/umap_popgen_review.bib"
```

```python
file = open(bibfilepath)

lines = file.readlines()
file.close()
```

```python
print(lines)
```

```python
# Fields to drop: copyright, abstract, publisher, note, language, month, 
new_lines = list()

for l in lines:
    print(l.strip())
    print()
    
    if not l.strip().startswith(("copyright","abstract","publisher","note","language","month","urldate")):
        new_lines.append(l)
```

```python
for l in new_lines:
    print(l)
```

```python
bibfilepath_out = "../ms_files/umap_popgen_review_trimmed.bib"
```

```python
f = open(bibfilepath_out, "w")
f.write("".join(str(i) for i in new_lines))
```
